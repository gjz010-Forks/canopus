```python
import sys
sys.path.append("..")

import canopus
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(2)
qc.iswap(0, 1)
qc = canopus.rebase_to_canonical(qc)
qc.draw(fold=100)
```

```python
qc_zz = canopus.rebase_to_zzphase(qc)
print(canopus.utils.is_equiv_unitary(canopus.utils.qc2mat(qc_zz), canopus.utils.qc2mat(qc)))

qc_zz.draw(fold=100)
```
