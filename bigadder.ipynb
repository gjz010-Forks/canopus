```python
import qiskit
from qiskit.transpiler import PassManager, passes, CouplingMap
```

```python
qc = qiskit.QuantumCircuit.from_qasm_file('./benchmarks/medium/bigadder_n18.qasm')
qc.num_qubits
```

```python
qc.draw(fold=10000)
```

```python
map = CouplingMap.from_grid(3, 6)
```

```python

pm = PassManager(passes.SabreLayout(map))
pm.run(qc)
```
