```python
from regulus import Circuit, gates
from cirq.vis.circuit_to_latex_render import render_circuit
import pytket
import qiskit
from regulus import mapping
from regulus.transforms import mirror
import rustworkx as rx
```

```python
circ = Circuit([
    gates.X.on(1, 0),
    gates.X.on(2, 0),
    gates.X.on(2, 1),
    gates.X.on(3, 0),
    gates.X.on(3, 1),
    gates.X.on(3, 2),
])
circ.to_qiskit().draw('mpl')
```

```python
# qiskit.transpile(circ.to_qiskit(), optimization_level=3, coupling_map=).draw('mpl')

mapping.sabre_by_qiskit(circ, rx.generators.path_graph(circ.num_qubits))[0].to_qiskit().draw('mpl')
```

```python
mirror.mirror_with_sabre(circ, rx.generators.path_graph(circ.num_qubits))[0].to_qiskit().draw('mpl')
```
